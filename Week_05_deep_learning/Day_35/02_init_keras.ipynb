```python
import pandas as pd
url = "https://raw.githubusercontent.com/Adityarajora/Complete-KNN-visualization/master/1.ushape.csv"
df = pd.read_csv(url, names=['X', 'Y', 'Class'])

df.head()
```

```python
X = df.iloc[:,0:-1].values
y = df.iloc[:,-1].values
import tensorflow 
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
model  = Sequential()
model.add(Dense(10,activation='relu',kernel_initializer='he_normal',input_dim = 2))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
history = model.fit(X,y,epochs=500,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'),clf=model,legend=2)
```
