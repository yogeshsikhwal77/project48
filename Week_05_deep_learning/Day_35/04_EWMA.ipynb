```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python


url = "https://raw.githubusercontent.com/swarnava-96/Time-Series/main/DailyDelhiClimateTest.csv"

df = pd.read_csv(
    url,
    usecols=["date", "meantemp"],  # Loads only these two columns
    parse_dates=["date"],  # Converts date string to datetime
)

df.head()
```

```python
plt.scatter(df[['date']],df['meantemp'],color='red')
```

```python
x1 = df['meantemp'].ewm(alpha=0.9).mean()
x2 = df['meantemp'].ewm(alpha=0.1).mean()
x1
```

```python
df['ewma1'] = x1
df['ewma2'] = x2
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='blue')
plt.plot(df['date'],x1,label='alpha=0.9',color='black')
plt.legend()
plt.show()
```

```python
plt.scatter(df['date'],df['meantemp'],color='blue')
plt.plot(df['date'],x2,label='alpha=0.9',color='black')
plt.legend()
plt.show()
```
