```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

# Generate simple dummy data for learning purposes
X = np.random.rand(1000, 10)
y = np.random.randint(0, 2, size=(1000, 1))
```

```python
model_m = Sequential()
model_m.add(Dense(32,activation='relu',input_dim = 10))
model_m.add(Dense(1,activation='sigmoid'))
```

```python
opt_m = SGD(learning_rate=0.01,momentum=0.9,nesterov=False)
model_m.compile(optimizer = opt_m,loss='binary_crossentropy',metrics=['accuracy'])
historym = model_m.fit(X,y,epochs=5,batch_size = 32)
```

```python
model_nag = Sequential()
model_nag.add(Dense(32,activation='relu',input_dim = 10))
model_nag.add(Dense(1,activation='sigmoid'))
opt_m = SGD(learning_rate=0.01,momentum=0.9,nesterov=True)
model_nag.compile(optimizer = opt_m,loss='binary_crossentropy',metrics=['accuracy'])
historynag = model_nag.fit(X,y,epochs=5,batch_size = 32)
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(8,5))
plt.plot(historym.history['accuracy'],label='Momentum',linewidth=2,color='blue')
plt.plot(historynag.history['accuracy'],label='nag',linewidth=2,color='red')
plt.legend()
plt.plot()
```
