```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()

or_data['input1'] = [1,1,0,0]
or_data['input2'] = [1,0,1,0]
or_data['outpur'] = [1,1,1,0]

and_data['input1'] = [1,1,0,0]
and_data['input2'] = [1,0,1,0]
and_data['outpur'] = [1,0,0,0]

xor_data['input1'] = [1,1,0,0]
xor_data['input2'] = [1,0,1,0]
xor_data['outpur'] = [0,1,1,0]
```

```python
and_data
```

```python
or_data
```

```python
xor_data
```

```python
from sklearn.linear_model import Perceptron
clf1 = Perceptron()
clf2 = Perceptron()
clf3 = Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
clf2.fit(and_data.iloc[:,0:2].values,or_data.iloc[:,-1].values)
clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)
```

```python
print(clf1.coef_)
print(clf1.intercept_)
```

```python
x = np.linspace(-1,1,5)
y = -x+1
plt.plot(x,y)
sns.scatterplot(x=and_data['input1'],y=and_data['input2'],hue=and_data['outpur'],s=200)
```

```python
x1 = np.linspace(-1,1,5)
y1 = -x+0.5
plt.plot(x1,y1)
sns.scatterplot(x=or_data['input1'],y=or_data['input2'],hue=or_data['outpur'],s=200)
```

```python
x = np.linspace(-1,1,5)
y = -x+1
plt.plot(x,y)
sns.scatterplot(x=xor_data['input1'],y=xor_data['input2'],hue=xor_data['outpur'],s=200)
```
