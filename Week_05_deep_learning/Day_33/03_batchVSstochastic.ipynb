```python
import numpy as np
import pandas as pd
import time
```

```python
url = 'https://raw.githubusercontent.com/Avik-Jain/100-Days-Of-ML-Code/master/datasets/Social_Network_Ads.csv'
df = pd.read_csv(url)
df.head()
```

```python
df = df[['Age','EstimatedSalary','Purchased']]
X = df.iloc[:,:2]
y = df.iloc[:,-1]
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10,activation='relu',input_dim = 2))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(X_scaled,y,epochs=50,batch_size=400,validation_split=0.2)
```

```python
model1 = Sequential()
model1.add(Dense(10,activation='relu',input_dim = 2))
model1.add(Dense(10,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',metrics=['accuracy'])
history1 = model1.fit(X_scaled,y,epochs=50,batch_size=1,validation_split=0.2)
```
