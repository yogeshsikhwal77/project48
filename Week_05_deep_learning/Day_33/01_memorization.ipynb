```python
import numpy as np
```

```python
def fib(n):
    if(n == 1 or n==0):
       return 1
    else :
       return fib(n-1) + fib(n-2)
```

```python
import time 
st = time.time()
fib(38)
print(time.time() - st)
```

```python
import time 
st = time.time()
fib(40)
print(time.time() - st)
```

```python
def fib1(n,d):
    if n in d:
       return d[n]
    else :
       d[n] = fib1(n-1,d) + fib1(n-2,d)
       return d[n]
```

```python
import time 
st = time.time()
d = {0:1,1:1}
fib1(40,d)
print(time.time() - st)
```
