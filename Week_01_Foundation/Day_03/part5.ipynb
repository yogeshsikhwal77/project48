## Numpy memory layout & contiguity

### 1.anatomy of an np.ndarray
- data pointers: ther memory address of very first byte
- shape: dimension of array ex (3,3)
- data pointers: ther memory address of very first byte


### 2.physical memory layouts
- C- contiguous (row-major):elements of same row are adjacent 
ex. [[1,2],[3,4]]    1D ram layout = [1,2,3,4]

- F- contiguous (collumn-major):elements of same columb are adjacent 
ex. [[1,2],[3,4]]    1D ram layout = [1,3,2,4]

### 3. Strides
strides are the secret of numpy speed
- let a (2,3) array of int64
- C-order strides: (24,8) means i want to jump 24 bites to reach next row and 8 bytes to reach he next column
- Zero-Copy Operations: When you transpose an array (arr.T), NumPy does not move the data in RAM. It simply reverses the strides to (8, 24).


### 4. contiguity matters
- CPU Cache Lines: CPUs load data from RAM in block chunks, not one number at a time.

- Contiguous Access (Fast): If you loop through an array in the exact order it is stored (e.g., row-by-row for C-contiguous), the CPU cache is perfectly utilized.

- Non-Contiguous Access (Slow): If you process a C-contiguous array column-by-column, the CPU has to constantly fetch new chunks from slow RAM (Cache Misses). This can make code 10x to 100x slower.

```python
import numpy as np
```

```python
arr = np.array([[1,2],[3,4]])
print(arr)
print(arr.flags['C_CONTIGUOUS'])
print(arr.flags['F_CONTIGUOUS'])
```

```python
arr_base = np.arange(6)
print(arr_base.flags['C_CONTIGUOUS'])

arr_reshaped = arr_base.reshape(2,3)
print(np.shares_memory(arr_base,arr_reshaped))
```

```python
grid = np.arange(16).reshape(4,4)
col_slice = grid[:,1]
print(col_slice)
print(col_slice.flags['C_CONTIGUOUS'])

slice_reshaped = col_slice.reshape(2,2)
print(np.shares_memory(col_slice,slice_reshaped))
```
