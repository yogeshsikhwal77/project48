## learn about numpy 

```python
import numpy as np
```

### creating array from list

```python
arr_1d = np.array([1,2,3,4,5])
print(arr_1d)

arr_2d = np.array([[1,2],[3,4]])
print(arr_2d)
```

## list vs numpy array

```python
py_list = [1,2,3]
print("list multipication: ",py_list)

np_array = np.array([1,2,3])
print("array multipication:",np_array)

import time
start = time.time()
py_list = [i*2 for i in range(10000000)]
print("list operation time: ",time.time() - start)
start = time.time()
np_array = np.arange(10000000) * 2
print("np_array operation time: ",time.time() - start)
```

### creating a array

```python
zeros = np.zeros((3,4))
print(zeros)

once = np.ones((2,3))
print(once)

full = np.full((2,3),7)
print(full)

random = np.random.random((2,3)) * 100
print(random)

sequence = np.arange(0,11, 2)
print(sequence)
```

### vector,matrix and tensor

```python
vector = np.array([1,2])
matrix = np.array([[1,2],[3,4]])
tensor = np.array([[[1,2],[3,4]],[[1,2],[3,4]]])
print(vector)
print(matrix)
print(tensor)
```

### array properties

```python
arr = np.array([[1,2,3],
               [4,5,6]])
print("shape ",arr.shape)
print("dimension ",arr.ndim) # important
print("size ",arr.size)
print("datatype",arr.dtype)
```

### array reshaping

```python
arr = np.arange(12)

reshaped = arr.reshape((2,6))
print(reshaped)

flattened = reshaped.flatten()
print(flattened)

# ravel ( return view instant of copy)
raveled = reshaped.ravel()
print(raveled)

#transpose
print(reshaped.T)
```
