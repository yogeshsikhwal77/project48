### Topic 20: Vector Norms (L^1 Manhattan, L^2 Euclidean, L^\infty maximum errors).


- $L^1$ Norm (Manhattan): $\sum |x_i|$. In ML, this is called Lasso Regularization. Because it measures strict grid-like distance, it tends to force the weights of irrelevant features to exactly $0$. It is a built-in feature selector.
- $L^2$ Norm (Euclidean): - $\sqrt{\sum x_i^2}$. In ML, this is called Ridge Regularization. It penalizes large weights exponentially but rarely forces them to $0$.- - -  - -
- L^\infty$ Norm (Max Norm): $\max(|x_i|)$. Used in specific deep learning constraints to ensure no single weight in a neural network grows disproportionately massive.

```python
v = np.array([-3,4,-5])

# Math: sqrt((-3)^2 + 4^2 + (-5)^2) = sqrt(9 + 16 + 25) = sqrt(50)
l2_norm = np.linalg.norm(v)

# Math: |-3| + |4| + |-5| = 3 + 4 + 5 = 12
l1_norm = np.linalg.norm(v,ord=1)

# Math: max(|-3|, |4|, |-5|) = 5
linf_norm = np.linalg.norm(v, ord=np.inf)

print(f"{l2_norm:.2f}")
print(f"{l1_norm:.2f}")
print(f"{linf_norm:.2f}")
```

```python
target = np.array([2.5, -1.0, 4.2, 0.5, -3.1])
dataset = np.random.randn(100000, 5)

reduced = dataset - target
distances1 = np.linalg.norm(reduced,axis= 1)
min_distance1 = np.argmin(distances1)
print(min_distance1)

distances2 = np.linalg.norm(reduced,ord=1,axis = 1)
min_distance2 = np.argmin(distances2)
print(min_distance2)
```

### Topic 21: Dot Products, geometric projections, and Cosine Similarity mechanics.


```python
v1=np.array([3,4])
v2=np.array([5,0])

# dot product
dot_product = np.dot(v1,v2)
print(dot_product)

#cosine imilarity 
# Formula: (A . B) / (|A| * |B|)
norm_v1 = np.linalg.norm(v1)
norm_v2 = np.linalg.norm(v2)

cosine_sim = dot_product/(norm_v1*norm_v2)
print(cosine_sim)

# GEOMETRIC PROJECTION (v1 onto v2)
g_proj_scaler = dot_product/np.dot(v2,v2)
g_proj_vector = g_proj_scaler *v2
print(g_proj_vector)
```

### Topic 22: Matrix Multiplication viewed as linear space transformations.


 - When you multiply $A \cdot B$, you are taking the space transformed by $B$ and transforming it again with $A$.
 
Layer Collapse

A basic neural network is just a sequence of matrix multiplications applied to an input vector $X$:

$$Output = W_3 \cdot (W_2 \cdot (W_1 \cdot X))$$

Because matrix multiplication is associative, you can group these weights:

$$Output = (W_3 \cdot W_2 \cdot W_1) \cdot X$$

$$Output = W_{fused} \cdot X$$

This is a profound realization. If you build a 1,000-layer neural network using only matrix multiplication, the entire network mathematically collapses into a single, simple transformation matrix. 

This is exactly why Neural Networks require non-linear **Activation Functions** (like ReLU or Sigmoid) between layers—to bend the space so it cannot be algebraically factored back down into a single linear matrix.

```python
X = np.array([
    [1,2,3], # x coordinate
    [1,1,1]  # y coordinate
])

theta = np.radians(90)
W1 = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

W2 = np.array([
    [1, 2],
    [0, 1]
])

transformed_sequential = W2 @ (W1 @ X)

W_fused = W2 @W1
transformed_fused = W_fused @ X

print("sequential and fused transformations mathematically identical")
print(np.allclose(transformed_sequential, transformed_fused))
```
