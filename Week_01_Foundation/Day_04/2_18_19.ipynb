### Topic 18: Coordinates, Scalars, Vectors, and Matrices

```python
scaler = 2.5
vector1 = np.array([5,3,4])
vector2 = np.array([2,7,3])
matrix_c = np.array([
    [0,0,0],
    [5,3,4],
    [2,7,3]
])

scaled_vector = scaler * vector1
added_vector = vector1 + vector2

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(0, 0, 0, color='black', s=50, label='Origin (0,0,0)')

ax.quiver(0,0,0,vector1[0],vector1[1],vector1[2], color='blue',arrow_length_ratio=0.1,label='Vector A')
ax.quiver(0,0,0,vector2[0],vector2[1],vector2[2], color='violet',arrow_length_ratio=0.1,label='scaled vector a 2.5x')

ax.set_xlim([-1, 8])
ax.set_ylim([-1, 8])
ax.set_zlim([-1, 15])
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title("3D Spatial Coordinates and Scalar Scaling")
ax.legend()

plt.show()

```

```python
scaler = 2

triangle_vertices = np.array([
    [1,1,0],
    [4,1,0],
    [2,4,0]
])

closed_triangle = np.vstack([triangle_vertices,triangle_vertices[0]])
scaled_triangle = closed_triangle * scaler
translation_vector = np.array([0,0,5])
translated_triangle = scaled_triangle + translation_vector



fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(closed_triangle[:,0],closed_triangle[:,1],closed_triangle[:,2], color='blue',marker='o',label='original triangle')
ax.plot(translated_triangle[:, 0], translated_triangle[:, 1], translated_triangle[:, 2], 
        color='red', marker='^', label='Scaled & Translated (z+5)')
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title("3D Spatial Coordinates and Scalar Scaling")
ax.legend()
plt.show()
```

```python
standard_points = np.random.rand(3, 1000)

new_basis = np.array([
    [1, 1, 0],
    [0, 2, 1],
    [1, 0, 1]
])




new_basis_points = np.linalg.inv(new_basis) @standard_points

# dimension collapse
new_basis_points[2, :] = 0
print(new_basis_points)

# muliply collapsed points by the original basis matrix to bring
reconstructed_points = new_basis @new_basis_points


print(reconstructed_points)




```
