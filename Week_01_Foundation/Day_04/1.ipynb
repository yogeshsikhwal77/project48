```python
import numpy as np
import matplotlib.pyplot as plt


def plot_transformation(matrix , title="linear transformation"):
    x= np.linspace(-5,5,10)
    y= np.linspace(-5,5,10)

    X,Y = np.meshgrid(x,y)

    points = np.vstack([X.flatten(), Y.flatten()])
    transformed_points = matrix@points
    X_trans = transformed_points[0,:]
    Y_trans = transformed_points[1,:]

    fig, ax = plt.subplots(1,2,figsize=(12,6))

    ax[0].scatter(X.flatten(),Y.flatten(),c='blue',alpha = 0.5,s=10)
    ax[0].set_title("original vector space")
    ax[0].grid(True)
    ax[0].axhline(0, color='red', linewidth=1)
    ax[0].axvline(0, color='pink', linewidth=1)
    ax[0].set_xlim([-10,10])
    ax[0].set_ylim([-10,10])

    ax[1].scatter(X_trans, Y_trans, c='red', alpha=0.5, s=10)
    ax[1].set_title(f"Transformed Space\n{title}")
    ax[1].grid(True)
    ax[1].axhline(0, color='black',linewidth=1)
    ax[1].axvline(0, color='black',linewidth=1)
    ax[1].set_xlim([-10, 10])
    ax[1].set_ylim([-10, 10])

    plt.show()

shear_matrix = np.array([[1,2],
                         [0,1]])

theta = np.radians(45)
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)],
])
print(rotation_matrix)

projection_matrix = np.array([
    [1,2],
    [3,4]
])

plot_transformation(shear_matrix, "Shear Transformation")
```
