### Topic 28: Matrix Diagonalization and the Spectral Theorem


ML Math Notes: Diagonalization & The Spectral Theorem

**Core Guarantee:** If a matrix is symmetric ($A = A^T$), it is diagonalizable as $A = Q \Lambda Q^T$, yielding **real eigenvalues** and **perfectly orthogonal eigenvectors**.

**1. The $O(N)$ Exponentiation Cheat Code**
* **Application:** Simulating future steps in Markov Decision Processes (RL) or Graph Neural Networks.
* **The Bottleneck:** Standard matrix multiplication for $A^{100}$ scales at a catastrophic $O(N^3)$.
* **The Bypass:** $A^{100} = Q \Lambda^{100} Q^T$. Because $\Lambda$ is a diagonal matrix, $\Lambda^{100}$ simply requires raising the scalar diagonal elements to the 100th power, reducing the core operation to $O(N)$.

**2. Feature Disentanglement (Zero Correlation)**
* **Application:** Principal Component Analysis (PCA) and dimensionality reduction.
* **The Mechanism:** The Spectral Theorem guarantees the eigenvectors of our Covariance matrix are orthogonal.
* **The ML Impact:** In data science, orthogonal vectors mathematically guarantee **zero linear correlation**. Projecting data onto these eigenvectors perfectly disentangles the features (PC1 and PC2 will have 0 correlation).

```python
np.random.seed(42)
A = np.random.randn(3,3)
symmetric_A = A @ A.T

eigenvalue,eigenvector = np.linalg.eigh(symmetric_A)

# np.diag() converts a 1D array into a 2D diagonal matrix
Lambda =  np.diag(eigenvalue) # lambda is diagonal matrix

# spectral theorem reconstruction
# A = Q @ Lambda @ Q^T
reconstructed_A = eigenvector @ Lambda @ eigenvector.T

is_perfect = np.allclose(symmetric_A,reconstructed_A)
print(is_perfect)
```

```python
np.random.seed(42)
A = np.random.randn(5, 5)
Transition_Matrix = A @ A.T # symmetric 

eigenvalue_1 , eigenvector_1 = np.linalg.eigh(Transition_Matrix)

Lambda_1 = np.diag(eigenvalue_1)
# Lambda_50 = np.linalg.matrix_power(Lambda,50) wrong 
Lambda_50 = Lambda_1 ** 50

reconstructed_matrix = eigenvector_1 @ Lambda_50 @ eigenvector_1.T

test = np.linalg.matrix_power(Transition_Matrix,50)

is_perfect_1 = np.allclose(test,reconstructed_matrix)
print(is_perfect_1)


```

### Topic 29: Singular Value Decomposition (SVD) Matrix Factorization

**Core Guarantee:** ANY matrix (even non-square) can be decomposed into $A = U \Sigma V^T$.
* $U$ and $V^T$ are orthogonal matrices (rotations/reflections).
* $\Sigma$ is a diagonal matrix of **Singular Values** (sorted descending).

**1. Truncated SVD (Image/Data Compression)**
* **Concept:** The largest singular values hold the most information. 
* **ML Application:** If you keep only the top $k$ singular values (and their corresponding vectors in $U$ and $V^T$), you can perfectly approximate the original matrix with a fraction of the memory.

**2. Recommender Systems (Latent Semantics)**
* **Concept:** Factoring a sparse User-Item matrix reveals hidden "latent" features (like movie genres or shopper habits).

```python
A = np.random.randn(400 , 10)

# 1.SVD 
U , singularValue , V_T = np.linalg.svd(A , full_matrices= False)
print("shape of U: ",U.shape)
print("shape of sigma (1D): ",singularValue.shape)
print("shape of V_T: ",V_T.shape)

# 2.reconstruct 
sigma_matrix = np.diag(singularValue)

reconstructed_2 = U @ sigma_matrix @ V_T
is_perfect_2 = np.allclose(A,reconstructed_2)
print(is_perfect_2)
```

```python


np.random.seed(42)
# A 500x500 "Image" Matrix
image_matrix = np.random.randn(500, 500)

U_3 , singularValue_3 , V_T_3 = np.linalg.svd(image_matrix,full_matrices=True)
slice_U_3 = U_3[: , :20]

slice_singularValue = singularValue_3[: 20]
diagonal_matrix = np.diag(slice_singularValue)
slice_V_T_3 = V_T_3[:20 , :]

compressed_image = slice_U_3 @ diagonal_matrix @ slice_V_T_3
print(compressed_image)
```
