```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# image convert to grayscale
img = Image.open("image.png").convert("L")
image_matrix = np.array(img , dtype=np.float64)

U , singular_matrix , V_T = np.linalg.svd(image_matrix,full_matrices=True)

k = 40
U_s = U[ : , :k]
singular_matrix_s = singular_matrix[ : k]
V_T_S = V_T[:k,:]
singular_matrix_s_dig = np.diag(singular_matrix_s)

image_matrix_R = U_s @ singular_matrix_s_dig @ V_T_S

# clipping
image_matrix_R = np.clip(image_matrix_R,0,255)
image_R = image_matrix_R.astype(np.uint8)

# visulixation
fig,axes = plt.subplots(1,2,figsize=(12,6))
axes[0].imshow(image_matrix,cmap="gray")
axes[0].set_title("original")
axes[1].imshow(image_matrix_R,cmap="gray")
axes[1].set_title("compressed")
plt.show()
```
