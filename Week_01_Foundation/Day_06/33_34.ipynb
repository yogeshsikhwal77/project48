### Topic 33: Computational Graphs & The Chain Rule
* **The Summation Rule:** When a tensor splits in the forward pass, its gradients **sum** in the backward pass. `grad_x = grad_path_1 + grad_path_2`.
* **Skip Connections (ResNets):** The mathematical engine of a skip connection is $F(x) + x$. The gradient backward is strictly $\nabla F(x) + 1$. That "$+ 1$" prevents the gradient from vanishing to zero.
* **Vectorized Backprop:** We do not calculate the full chain rule from scratch at every layer. We take the gradient from the layer ahead of us (`upstream_grad`), and matrix-multiply it by our local derivative (`local_grad`).

```python
x= np.array([[2.0],[3.0]])

#forward pass
y = x**2 
z = x
L =np.sum(y+z)
print(f"Loss: {L}")

#backwared pass 
dL_dy = np.ones_like(y)
dL_dz = np.ones_like(z)

dy_dx = 2 * x
dz_dx = 1

grad_x = (dL_dz * dy_dx) + (dL_dz * dz_dx)
print(grad_x)
```

```python
X = np.array([[1.0], [2.0]]) 
W = np.array([[0.5, 1.0],
              [-1.0, 2.0]])

# Forward Pass (Do not change this)
H = W @ X
A = H**2
B = 3 * X
Y = A + B
L = np.sum(Y)
print(L)

# Backward Pass
dL_dA = np.ones_like(A)
dL_dB = np.ones_like(B)

grid_X_path_B = dL_dB * 3
dL_dH = dL_dA*(2*H)
grid_X_path_A = W.T @ dL_dH

grad_x_total = grid_X_path_A + grid_X_path_B
print(grad_x_total)




```

### Topic 34: The Gradient Vector & Local Ascent/Descent
* **Directional Guarantee:** The gradient vector $\nabla f$ points in the direction of the steepest local increase of the function.
* **Gradient Descent Formula:** $W \leftarrow W - \alpha \nabla_W L$. Moving against the gradient minimizes loss.
* **Vector Mechanics:** In NumPy, optimization means evaluating the gradient vector across a batch of data points cleanly without loops, updating the parameter space in a single vectorized step.
$D_u f = \nabla f \cdot \mathbf{u} = \|\nabla f\| \|\mathbf{u}\| \cos(\theta)$
$$W_{new} = W_{old} - \alpha \nabla_W L$$
where $\alpha$ is the learning rate (step size).

```python
# L(w1, w2) = w1^2 + 3 * w2^2
# Global minimum is at (0, 0)
def compute_loss(W):
    return W[0,0]**2 + 3 *W[1,0]**2

def compute_gradient(W):
    # Analytical gradient vector: [dL/dw1, dL/dw2]^T
    # dL/dw1 = 2*w1,  dL/dw2 = 6*w2
    grad_w1 = 2*W[0,0]
    grad_w2 = 6*W[1,0]
    return np.array([[grad_w1],[grad_w2]])


W = np.array([[4.0],[2.0]])
learning_rate = 0.1

print("initial weights: ",W)
print("initial lost: ",compute_loss(W))

grad = compute_gradient(W)
print("gradient vector" ,grad)

W = W - learning_rate * grad

print("weights after 1 step: ",W)
print("loss after step 1: ",compute_loss(W))

```

```python
def compute_loss(W):
    return 0.5*(W[0,0]**2) + 2 *W[1,0]**2

def compute_gradient(W):
    grad_w1 = W[0,0]
    grad_w2 = 4*W[1,0]
    return np.array([[grad_w1],[grad_w2]])
    

W = np.array([[4.0],[4.0]])
learning_rate = 0.1
print("initial weights: ",W)
print("initial lost: ",compute_loss(W))

for i in range(10):
    grad = compute_gradient(W)
    
    W = W - learning_rate * grad
    
    print(f"loss after step {i}: ",compute_loss(W))
```
