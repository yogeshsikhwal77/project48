```python
r_text = """
artificial intelligence is the simulation of human intelligence by machines
machine learning is a powerful branch of artificial intelligence
deep learning uses multi layer neural networks to analyze complex data
neural networks are inspired by the biological structure of the human brain
early neural networks were simple perceptrons with limited computing power
modern deep learning models require immense computational resources and data
natural language processing focuses on the interaction between computers and human language
the goal of natural language processing is to read decipher and understand text
early language models relied on statistical methods and simple frequency counts
recurrent neural networks introduced the ability to process sequential data over time
recurrent neural networks suffered from the vanishing gradient problem during training
long short term memory networks solved the vanishing gradient problem effectively
long short term memory networks use specialized gates to remember and forget information
the input gate controls how much new data flows into the memory cell
the forget gate decides what information should be discarded from the cell state
the output gate determines what the next hidden state should be
for many years long short term memory models dominated natural language processing tasks
scientists used these networks for machine translation speech recognition and text generation
despite their success sequential processing made training large recurrent networks very slow
in twenty seventeen researchers introduced the revolutionary transformer architecture
the transformer architecture relies entirely on self attention mechanisms without recurrent layers
self attention allows models to look at all words in a sentence simultaneously
this parallel processing speed enabled the creation of massive foundational language models
today large language models can generate human like text across diverse topics
fine tuning adapts pre trained models to follow specific instructions safely
retrieval augmented generation allows language models to access external knowledge bases
as artificial intelligence evolves efficient architectures will continue to emerge
understanding classical recurrent networks remains essential for modern machine learning engineers
the journey from simple perceptrons to massive language models is truly remarkable
every new architecture builds upon the mathematical foundations of its predecessors
"""
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([r_text])
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in r_text.split('\n'):
    tokenizer_sentence = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1,len(tokenizer_sentence)):
        input_sequences.append(tokenizer_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen = max_len,padding='pre')
```

```python
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=199)
y
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,  LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(199,100,input_length=12))
model.add(LSTM(150))
model.add(Dense(199,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
text = "every"
for i in range(6):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  #padding
  padded_text_token = pad_sequences([token_text],maxlen=12,padding='pre')
  #predict
  import numpy as np
  t = np.argmax(model.predict(padded_text_token))
  for word,index in tokenizer.word_index.items():
      if index == t:
          text = text + " " + word
          print(text)
```
