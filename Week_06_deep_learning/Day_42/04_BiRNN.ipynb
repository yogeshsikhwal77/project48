```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN,Dense,LSTM,GRU,Bidirectional
```

```python
num_words = 10000
(X_train,y_train),(X_test,y_test) = imdb.load_data(num_words=num_words)

x_train = pad_sequences(X_train,maxlen=100,padding='post',truncating='post')
x_test = pad_sequences(X_test,maxlen=100,padding='post',truncating='post')
```

```python
Embedding_dim = 32

model = Sequential([
    Embedding(input_dim=num_words,output_dim=Embedding_dim,input_length=100),
    SimpleRNN(5),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
Embedding_dim = 32

model = Sequential([
    Embedding(input_dim=num_words,output_dim=Embedding_dim,input_length=100),
    Bidirectional(SimpleRNN(5)),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
Embedding_dim = 32

model = Sequential([
    Embedding(input_dim=num_words,output_dim=Embedding_dim,input_length=100),
    Bidirectional(LSTM(5)),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
Embedding_dim = 32

model = Sequential([
    Embedding(input_dim=num_words,output_dim=Embedding_dim,input_length=100),
    Bidirectional(GRU(5)),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```
