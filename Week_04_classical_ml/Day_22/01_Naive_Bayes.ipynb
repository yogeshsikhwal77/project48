## Problem 1
* Outlook=sunny temp=hot humidity=high wind=weak
- play or not

### Solution 
- P(Yes | sunny,hot,high,weak) = P(sunny | yes) * P(Hot | yes) * P(High | yes) * P(week | yes) * P(yes)
- P(no | sunny,hot,high,weak) = P(sunny | no) * P(Hot | no) * P(High | no) * P(week | no) * P(No)
* compare and decide using the maximum a posteriori rule 

## Problem 2

- play or not

### Solution 
- P(Yes | overcast,cold,low,weak) = P(overcast | yes) * P(cold | yes) * P(Low | yes) * P(week | yes) * P(yes)
- P(no | overcast,cold,low,weak) = P(overcast | no) * P(cold | no) * P(Low | no) * P(week | no) * P(No)
* compare and decide using the maximum a posteriori rule
