```python
import numpy as np
from sklearn.datasets import load_iris, load_diabetes
from sklearn import tree
import dtreeviz
```

```python
clas = tree.DecisionTreeClassifier(max_depth=3, random_state=42)  
iris = load_iris()
X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clas)
```

## 1.Classification

```python
import os

# The standard installation path for Graphviz on Windows
graphviz_path = r'C:\Program Files\Graphviz\bin'

# Append it to the system PATH for this Python session
os.environ["PATH"] += os.pathsep + graphviz_path
viz_model = dtreeviz.model(
    clas,
    X_train=X_train,
    y_train=y_train,
    feature_names=iris.feature_names,
    class_names=["setosa", "versicolor", "virginica"],
    target_name="iris"
)

viz_model.view()
```

## 10. Univariate Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
url = "https://raw.githubusercontent.com/parrt/dtreeviz/master/data/cars.csv"

df_cars = pd.read_csv(url)
df_cars.head()
```

```python
X_cars = df_cars[['WGT']]
y_cars = df_cars['MPG']
```

```python
dt_univar = DecisionTreeRegressor(max_depth=3,criterion="absolute_error")
dt_univar.fit(X_cars,y_cars)

viz_univar = dtreeviz.model(
    dt_univar,
    X_train=X_cars,
    y_train=y_cars,
    feature_names=['WGT'],
    target_name='MPG'
)
viz_univar.rtree_feature_space(features=['WGT'])
plt.show()
```

## 11. 3-D Bivariate Regression

```python
X_bivar = df_cars[['WGT','ENG']]
y_bivar = df_cars['MPG']
dt_bivar = DecisionTreeRegressor(max_depth=3,criterion="absolute_error")
dt_bivar.fit(X_bivar,y_bivar)
viz_bivar = dtreeviz.model(
    dt_bivar,
    X_train=X_bivar,
    y_train=y_bivar,
    feature_names=['WGT','ENG'],
    target_name='MPG'
)
viz_bivar.rtree_feature_space3D(features=['WGT','ENG'],elev=20,azim=25,dist=8.2,fontsize=14)
plt.show()
```
